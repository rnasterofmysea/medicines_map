크롤링 되는지 여부 판단

```python
import requests

def check_crawling_restrictions(url):
    try:
        response = requests.get(url)
        if response.status_code == 200:
            print("크롤링이 허용되는 웹사이트입니다.")
        elif response.status_code == 403:
            print("접근이 금지된 웹사이트입니다. 크롤링이 제한되었습니다.")
        else:
            print("웹사이트에 접속할 수 없습니다.")
    except requests.exceptions.RequestException as e:
        print("오류가 발생하였습니다:", e)

# 웹사이트 URL
url = "https://www.e-gen.or.kr/egen/search_pharmacy.do?searchType=map"

# 크롤링 제한 여부 확인
check_crawling_restrictions(url)
```

```python
import requests
from bs4 import BeautifulSoup
import certifi

def scrape_pharmacy_data():
    url = "https://www.e-gen.or.kr/egen/search_pharmacy.do?searchType=map"

    try:
        # 인증서 파일 로드
        cafile = certifi.where()

        # 웹사이트에 GET 요청 보내기 (SSL 인증서 검증)
        response = requests.get(url, verify=cafile)
        
        if response.status_code == 200:
            # 응답 데이터를 BeautifulSoup을 사용해 파싱하기
            soup = BeautifulSoup(response.content, "html.parser")
            
            # 크롤링할 내용을 선택자를 사용하여 추출하기
            # 예시: 약국 이름 추출
            pharmacy_names = soup.select(".tbl_type01 td:first-child a")
            
            # 추출한 데이터 처리 또는 출력하기
            for name in pharmacy_names:
                print(name.text.strip())
                
        else:
            print("웹사이트에 접속할 수 없습니다.")
            
    except requests.exceptions.RequestException as e:
        print("오류가 발생하였습니다:", e)

# 약국 데이터 크롤링 실행
scrape_pharmacy_data()
```
